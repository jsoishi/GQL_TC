```python
import dedalus.public as de
import numpy as np
import matplotlib.pyplot as plt
```

Create a Dedalus basis (Chebyshev) and a domain (one-D here)

```python
nr = 32
R1 = 7
R2 = 8

r = de.Chebyshev('r', nr, interval=(R1,R2))
d = de.Domain([r], grid_dtype=np.float)
```

```python
v_r_DNS = d.new_field()
```
